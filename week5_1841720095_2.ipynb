**Import Libraries and Dataset**

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
weather = pd.read_csv("/content/drive/MyDrive/TDATA/dataset/weatherlondon.csv")
weather
```

**Simple Pandas Graph**

```python
weather.plot(y='Tmax', x='Month')
plt.show
```

**Multiple Line Plots**

```python
weather.plot(y=['Tmax','Tmin'],x='Month')
plt.show
```

**Bar Chart**

```python
weather.plot(kind='bar', y='Rain', x='Month')
plt.show()
```

**Horizontal Bar Chart**

```python
weather.plot(kind='barh', y='Rain', x='Month')
plt.show()
```

**Pie Chart**

```python
weather.plot(kind='pie', y='Sun')
plt.show()
```

**Scatterplot**

```python
weather.plot(kind='scatter', x='Sun', y='Rain')
plt.show()
```

**Multiple Chart**

```python
weather.plot(y=['Tmax', 'Tmin', 'Rain', 'Sun'], subplots=True, layout=(2,2), figsize=(10,5))
plt.show()
```

**Simply Saving Chart**

```python
weather.plot(kind='pie', y='Rain', legend=False)
plt.show()
plt.savefig('pie.png')
```
